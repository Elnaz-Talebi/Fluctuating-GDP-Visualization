# Countries by GDP growth percentage 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
GDP_Growth_copy = pd.read_csv("GDP-Growth_covid1.csv")
GDP_Growth = GDP_Growth_copy.copy()
GDP_Growth
```

```python
GDP_Growth = GDP_Growth.drop(columns=["Last","Previous","Unit"])
```

```python
GDP_Growth.rename(columns = {'Reference':'Year','GDP_growth_percentage':'GDP_Growth_Percentage','GDP_growth_percentage.1':'GDP_Growth_State'}, inplace = True)
GDP_Growth
```

```python
GDP_Growth['GDP_Growth_Percentage'].dtypes
```

```python
Negative_GDP_Growth = GDP_Growth[GDP_Growth["GDP_Growth_Percentage"] < 0]
Positive_GDP_Growth = GDP_Growth[GDP_Growth["GDP_Growth_Percentage"] > 0]

ax = Negative_GDP_Growth.plot.bar(x='Country Name', y='GDP_Growth_Percentage')
ax = Positive_GDP_Growth.plot.bar(x='Country Name', y='GDP_Growth_Percentage')
```

```python
color = ['r' if y<0 else 'g' for y in GDP_Growth["GDP_Growth_Percentage"]]
ax = GDP_Growth.plot.bar(x='Country Name', y='GDP_Growth_Percentage', 
                         color=color, figsize=(18,7), legend=False, title='Countries by GDP growth percentage'
                        ,xlabel='Country Name',ylabel='GDP Growth Percentage' )
```

```python
GDP_Growth
```

```python
GDP_Growth_groupby_growth = GDP_Growth.groupby('GDP_Growth_State')['Country Name'].count()

ax = GDP_Growth_groupby_growth.plot.bar(x='GDP_Growth_State',figsize=(8,7), legend=False, title='GDP Growth State count'
                        ,xlabel='Growth State',ylabel='Country Count')
```
