# Change oriantion of the table

```python
gdp1 = GDP_original.copy()
GDP_covid = pd.read_csv("gdp_world_covid_2020_new.csv")
gdp_covid_copy = GDP_covid.copy()
```

```python
gdp1
```

```python
simple_gdp = gdp1.drop('Continent', inplace=True, axis=1)
print(simple_gdp)
```

```python
simple_gdp = gdp1.drop('Country Code', inplace=True, axis=1)
display(simple_gdp)
```

```python
gdp1
```

```python
melted_gdp1 = pd.melt(gdp1, id_vars=["Country Name"], var_name="Year", value_name="GDP")
display(melted_gdp1.head())

pivoted_gdp1 = melted_gdp1.pivot(index='Year', columns='Country Name', values='GDP')
display(pivoted_gdp1.head())
```

```python
top_10_countries_new = pivoted_gdp1.loc['2019':].T.dropna().sort_values(by='2019', ascending=False).head(10).index.tolist()
```

```python
top_10_countries_new
```

```python
# display(pivoted_gdp1['Year'])
pivoted_gdp1 = pivoted_gdp1.reset_index()

pivoted_gdp1['Year'] = pd.to_datetime(pivoted_gdp1['Year'])
pivoted_gdp1 = pivoted_gdp1.set_index('Year')
pivoted_gdp1.tail()
```

```python
temp = pivoted_gdp1[top_10_countries_new].reset_index().fillna('')
temp.plot_bokeh.area(
    x="Year",
    stacked=True,
    legend="top_left",
    #colormap=["brown", "orange", "black", "grey", "blue", "green"],
    title="GDP Growth of the top 10 Countries",
    ylabel="USD",
    ylim=(0, 6e+13))
```

```python
gdp_covid_copy['GDP_COVID'] = gdp_covid_copy['Last'] * 1000000000
gdp_covid_copy
```

```python
Final_lst_gdp_covid_copy = gdp_covid_copy.drop(['Last', 'Previous','Reference','Unit'], axis = 1)
Final_lst_gdp_covid_copy
```

# Merge gdp & Covid

```python
mrg_gdp_covid = pd.merge(gdp, Final_lst_gdp_covid_copy[["Country Name","GDP_COVID"]],on="Country Name",how="left")
mrg_gdp_covid
```

```python
Top_10_countries_GDP_COVID = display(mrg_gdp_covid[['Country Name', 'Country Code', 'GDP_COVID']].sort_values('GDP_COVID', ascending=False).head(10))
Top_10_countries_GDP_COVID
```

```python
Top_3_gdp_covid = mrg_gdp_covid[mrg_gdp_covid['Country Name'].isin([
     'United States',
     'India',
     'Japan',
     'United Kingdom','France','India','Italy','Canada','Korea, Rep.','Australia'
    ])].fillna(0)
Top_3_gdp_covid.plot_bokeh.pie(
    x="Country Name",
    y="GDP_COVID",
    title="Top 10 Countries by GDP during Covid",
    )
```
